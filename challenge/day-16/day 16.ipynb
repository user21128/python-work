#### Q5. Count vowels in a string.

```python
s=input("Enter a string: ")
vowels="aeiouAEIOU"
count=0
for c in s:
    if c in vowels:
        count+=1
print(f"The number of vowels in the string is: {count}")
```

#### Q9. Write a function to check prime number.

```python
def prime(n):
    if n>1:
        for i in range(2,n):
            if n%i==0:
                return False
        else:
            return True
    else:
        return False
n=int(input("Enter a number: "))
print(f"is {n} prime no:", prime(n))
```

#### Q10. Find key with maximum value.

```python
d = {"A": 10, "B": 50, "C": 30}

max_key = max(d, key=d.get)

print("Max key:", max_key)
```
