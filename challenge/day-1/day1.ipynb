```python
def merge(dict1, dict2):
    merged = dict1.copy()
    merged.update(dict2)
    return merged
dict1 = {"a": 1, "b": 2}
dict2 = {"c": 3, "d": 4}
merged = merge(dict1, dict2)
print(merged)

```
