#### Q1. Count frequency of words in sentence.

```python
s = input("Enter a sentence: ").lower()

words = s.split()
freq = {}

for w in words:
    freq[w] = freq.get(w, 0) + 1

print(freq)
```

#### 2. Remove duplicate characters from string.

```python
s=input("Enter a string: ")
result=""
for c in s:
    if c not in result:
        result+=c
print(result)
```

#### 3. Find all pairs in list with given sum.

```python
lst=[1,2,3,4,5,6,7,8,9]
target=int(input("Enter target sum: "))
for i in range(len(lst)):
    for j in range(i+1,len(lst)):
        if lst[i]+lst[j]==target:
            print(f"Pair: ({lst[i]}, {lst[j]})")
```

#### Q4. Demonstrate continue statement.

```python
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

#### Q6. Count uppercase and lowercase letters.

```python
s=input("Enter a string: ")

upper=lower=0
for c in s:
    if c.isupper():
        upper+=1
    elif c.islower():
        lower+=1
print(f"Uppercase letters: {upper}")
print(f"Lowercase letters: {lower}")
```

#### Q7. Infinite loop until user types "exit"

```python
while True:
    print("This will keep running until you type 'exit'.")
    text = input("Enter something: ")
    if text == "exit":
        break
```

#### 8. Find the greatest among three numbers assuming no two values are same. 

```python
a=int(input("Enter a no:"))
b=int(input("Enter a sec no:"))
c=int(input("Enter a third no:"))

if a>b and a>c:
    print(f"{a} is greater no")
elif b>a and b>c:
    print(f"{b} is greater no")  
elif c>a and c>b:
    print(f"{c} is greater no") 
```

#### Q9. Check palindrome string.

```python
m=input("Enter a string: ")
if m==m[::-1]:
    print(f"{m} is a palindrome string.")
else:
    print(f"{m} is not a palindrome string .")
```
