#### Q4. Write a function to count vowels in a string.

```python
def count_vowels(s):
    count=0
    for c in s.lower():
        if c in "aeiou":
            count+=1
    return count

s=input("Enter a string: ")
print("Number of vowels =", count_vowels(s))
```

#### Q5. Find all Armstrong numbers between 1–500

```python


armstrong_numbers = []
for num in range(1, 501):
    order = len(str(num))
    sum = 0
    temp = num
    while temp > 0:
        digit = temp % 10
        sum += digit ** order
        temp //= 10
    if num == sum:
        armstrong_numbers.append(num)

print("Armstrong numbers between 1 and 500:", armstrong_numbers)
```
