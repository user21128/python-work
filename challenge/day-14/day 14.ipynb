#### Q1. Write a recursive function for Fibonacci.

```python
def fib(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b

n=int(input("Enter a number: "))
fib(n)
```

#### Q7. Student Search System

```python
students = {
    "Alice": {"age": 20, "course": "CS"},
    "Bob": {"age": 22, "course": "Math"},
    "Charlie": {"age": 21, "course": "Physics"}
}
name = input("Enter student name: ")
if name in students:
    print(f"Age: {students[name]['age']}, Course: {students[name]['course']}")
else:
    print("Student not found")
```

#### Q9. Write a function to find factorial.

```python
def factorial(n):
    fact=1
    for i in range(1,n+1):
        fact*=i
    return fact
n=int(input("Enter a number: "))
print("Factorial =", factorial(n))
```

#### Q10. Count words starting with vowel.

```python
s=input("Enter a sentence: ").lower()
words=s.split()
count=0
for w in words:
    if w[0] in "aeiou":
        count+=1
print("Number of words starting with vowel:", count)
```
