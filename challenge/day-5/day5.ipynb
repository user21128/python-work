```python
# building a program to verify thr following data from a txt file (name, age, phone) using for loop and if statement
with open('data.txt', 'r') as file:
    for line in file:
        name, age, phone = line.strip().split(',')
        if name and age and phone:
            print(f"Name: {name}, Age: {age}, Phone: {phone}")
        else:
            print("Invalid data format.")
            break
```

```python
# lets try to retrive the data from the text file only after inputting the correct name else it will print invalid date
with open('data.txt', 'r') as file:
    name_to_find = input("Enter the name to find: ")
    for line in file:
        name, age, phone = line.strip().split(',')
        if name == name_to_find:
            print(f"Name: {name}, Age: {age}, Phone: {phone}")
            break
    else:
        print("Invalid data format.")
```

```python
# lets try to add new data into our text file without any help of the libraries
with open('data.txt', 'a') as file:
    new_name = input("Enter the name to add: ")
    new_age = input("Enter the age to add: ")
    new_phone = input("Enter the phone number to add: ")
    file.write(f"{new_name},{new_age},{new_phone}\n\n")
    
```

```python
#lets try to delete a specific line from the text file by inputting the name
with open('data.txt', 'r') as file:
    lines = file.readlines()
    name_delete = input("Enter the name to delete: ")
    with open('data.txt', 'w') as file:        
        for line in lines:
            name, age, phone = line.strip().split(',')
            if name != name_delete:
                file.write(line)
                break
```

```python
with open('data.txt', 'r') as file:
    lines = file.readlines()
    name_replace = input("Enter the name to replace: ")
    new_name = input("Enter the new name: ")
    new_age = input("Enter the new age: ")
    new_phone = input("Enter the new phone number: ")
    with open('data.txt', 'w') as file:        
        for line in lines:
            name, age, phone = line.strip().split(',')
            if name == name_replace:
                file.write(f"{new_name},{new_age},{new_phone}\n\n")
                break
            else:
                file.write(line)
```
