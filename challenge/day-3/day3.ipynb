# day 3 of this challenge lets get started

```python
def main():
    return {
        "x": {"name": "x", "age": 30, "city": "city1"},
        "y": {"name": "y", "age": 25, "city": "city2"},
        "z": {"name": "z", "age": 35, "city": "city3"}
    }

dic = main()
```

```python
for k in dic:
    print(k)
```

```python
for v in dic.values():
    print(v)
```

```python
for k, v in dic.items():
    print(f"{k}: {v}")
```

```python
for p in dic:
    for detail in dic[p]:
        print(f"{p} {detail}: {dic[p][detail]}")
```

```python
keys = [k for k in dic]
print(keys)
```

```python
vals = [v for v in dic.values()]
print(vals)
```

```python
items = [item for item in dic.items()]
print(items)
```
