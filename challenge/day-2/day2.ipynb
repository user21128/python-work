```python
dic = {
    "x": {"name": "x", "age": 39, "city": "city1"},
    "y": {"name": "y", "age": 265, "city": "city2"},
    "z": {"name": "z", "age": 35, "city": "city3"}
}

for p, d in dic.items():
    print(f"{p}: {d['name']} is {d['age']} years old and lives in {d['city']}.")
```

```python
print(dic["x"]["age"])
```

```python
dic["x"]["age"] = 31
print(dic["x"])
```

```python
dic["w"] = {"name": "w", "age": 40, "city": "city4"}
print(dic)
```
