#### Q1. Print numbers in triangle form:</h4><h5>1<br>23<br>456<br>78910</h5>

```
num = 1
for i in range(1, 5):
    for j in range(i):
        print(num, end="")
        num += 1
    print()
```

#### Q2. Take a string and print frequency of vowels using dictionary.

```
s=input("Enter a string: ").lower()
freq={"a":0,"e":0,"i":0,"o":0,"u":0}

for ch in s:
    if ch in freq:
        freq[ch] += 1

print(freq)
```

#### Q3. Check if two strings are anagrams.

```
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")
if sorted(s1) == sorted(s2):
    print("Anagrams")
else:
    print("Not anagrams")
```

#### Q9. Print all prime numbers between 1 to 50.

```
for num in range(1,51):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num,end=" ")
```

#### 10. Scan n values in range 0-3 and print the number of times each value has occurred.

```
n=int(input("enter a number: "))
freq=[0,0,0,0]

for i in range(n):
    num=int(input("enter a number(0-3): "))
    if 0<=num<=3:
        freq[num]+=1
for i in range(4):
    print(i,"occurred", freq[i],"times")
```
