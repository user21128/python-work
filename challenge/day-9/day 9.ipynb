#### Q1. Write a lambda function to find square.

```python
square = lambda x: x * x

print(square(int(input("Enter a number: "))))
```

#### Q2. Check palindrome number.

```python
n=int(input("Enter a number: "))
temp=n
rev=0
while temp>0:
    digit=temp%10
    rev=rev*10+digit
    temp//=10
if rev==n:
    print(f"{n} is a palindrome number.")
else:
    print(f"{n} is not a palindrome number.")
```

#### Q3. Find factorial of a number.

```python
n=int(input("Enter a number: "))
fact=1
for i in range(1,n+1):
    fact*=i
print(f"Factorial of {n} is {fact}")
```

#### Q5. Remove duplicate characters from string.

```python
s=input("Enter a string: ")
result=""
for c in s:
    if c not in result:
        result+=c
print(result)
```

```python
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")
if sorted(s1) == sorted(s2):
    print("Anagrams")
else:
    print("Not anagrams")
```
