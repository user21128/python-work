```python
# content counting
with open('names.txt', 'r') as file:
    content = file.read()
    word_count = len(content.split())
    print(f'Word count: {word_count}')
```

```python
#count words starting with vowels
with open('names.txt', 'r') as file:
    content = file.read()
    words = content.split()
    vowel_count = sum(1 for word in words if word[0].lower() in 'aeiou')
    print(f'Words starting with a vowel: {vowel_count}')
```

```python
#find longest name
with open('names.txt', 'r') as file:
    content = file.read()
    words = content.split()
    longest_name = max(words, key=len)
    print(f'Longest name: {longest_name}')
```

```python
#find the max number in a file
with open('text.txt', 'r') as file:
    content = file.read()
    numbers = [int(num) for num in content.split() if num.isdigit()]
    max_number = max(numbers)
    print(f'Max number: {max_number}')
```

```python
#find average of all numbers
with open('text.txt', 'r') as file:
    content = file.read()
    numbers = [int(num) for num in content.split() if num.isdigit()]
    average = sum(numbers) / len(numbers) if numbers else 0
    print(f'Average number: {average}')
```

```python
#count no of numbers greater than 100
with open('text.txt', 'r') as file:
    content = file.read()
    numbers = [int(num) for num in content.split() if num.isdigit()]
    count_greater_than_100 = sum(1 for num in numbers if num > 100)
    print(f'Numbers greater than 100: {count_greater_than_100}')
```

```python
#display detailsof all cities from city.txt
with open('city.txt', 'r') as file:
    content = file.read()
    cities = content.splitlines()
    for city in cities:
        print(city)
```

```python
# display city with population greater than 1 million
with open('city.txt', 'r') as file:
    content = file.read()
    cities = content.splitlines()
    for city in cities:
        if int(city.split(',')[1]) > 10:
            print(f'{city.split(",")[0]} has a population greater than 1 million.')
```

```python
# sum of areas of all cities
with open('city.txt', 'r') as file:
    content = file.read()
    cities = content.splitlines()
    total_area = sum(int(city.split(',')[2]) for city in cities)
    print(f'Total area of all cities: {total_area}')
```

```python
# multiple exceptions for a file handling program
with open('data.txt', 'r') as file:
    try:
        content = file.read()
        numbers = [int(num) for num in content.split() if num.isdigit()]
        average = sum(numbers) / len(numbers) if numbers else 0
        print(f'Average number: {average}')
    except ZeroDivisionError:
        print('Error: No numbers to calculate average.')
    except FileNotFoundError:
        print('Error: File not found.')
```
