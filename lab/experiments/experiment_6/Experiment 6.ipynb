```python
student={}
n=int(input("enter how many students:"))
for i in range(1,n+1):
    name=input("enter student name:")
    city=input("enter city name:")
    student[name]=city
print(student)
print("\nStudents name are")
for name in student.keys():
    print(name)
print("\nCities are")
for city in student.values():
    print (city)
print("\nStudent and their cities are")
for name,city in student.items():
    print(name,"lives in", city)

city_count={}
for city in student.values():
    if city in city_count:
        city_count[city]+=1
    else:
        city_count[city]=1

print("\nNumber of students in each city")
for city,count in city_count.items():
    print(city,"has" ,count,"students")
```

```python
movie={}
n=int(input("enter number of movies:"))
for i in range(1,n+1):
    name=input("enter movie name:")
    year=int(input("enter release year:"))
    director_name=input("enter director name:")
    production_cost=int(input("enter production cost:"))
    collection=int(input("enter collection:"))
    movie[name]=(year, director_name, production_cost, collection)
print("\nMovies and their details are:")
for name,details in movie.items():
    print(f"Movie: {name}")
    print(f"Year: {details[0]}")
    print(f"Director: {details[1]}")
    print(f"Production Cost: {details[2]}")
    print(f"Collection: {details[3]}")
    print()
print("\nMovies released before 2015:")
for name, details in movie.items():
    if details[0] < 2015:
        print(f"Movie: {name}")
        print(f"Year: {details[0]}")
        print()
print("\nMovies that made a profit:")
for name,details in movie.items():
    if details[3]>details[2]:
        print(f"Movie: {name}")
        print(f"Collection: {details[3]}")
        print(f"Production Cost: {details[2]}")
        print(f"Profit: {details[3]-details[2]}")
        print()
print ("\nMovies directed by a particular director:")
director = input("Enter the director's name: ")
for name, details in movie.items():
    if details[1] == director:
        print(f"Movie: {name}")
        print()
```
