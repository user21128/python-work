Find a factorial of given number.  

```python
#factorial of a number using recursiondef factorial(n):
input_number = int(input("Enter a number: "))
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
print(f"The factorial of {input_number} is: {factorial(input_number)}")
```

Find whether the given number is Armstrong number. 

```python
input_number = int(input("Enter a number: "))
def armstrong_number(n):
    sum_of_digits = 0
    temp = n
    while temp > 0:
        digit = temp % 10
        sum_of_digits += digit ** 3
        temp //= 10
    return sum_of_digits == n
print(f"{input_number} is an Armstrong number: {armstrong_number(input_number)}")
```

Print Fibonacci series up to given term. 

```python
input_number = int(input("Enter a number: "))
def fibonacci(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    else:
        sequence = [0, 1]
        for i in range(2, n):
            next_number = sequence[i - 1] + sequence[i - 2]
            sequence.append(next_number)
        return sequence
print(f"The first {input_number} numbers in the Fibonacci sequence are: {fibonacci(input_number)}")
```

Write a program to find if given number is prime number or not. 

```python
number = int(input("Enter a number: "))
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
print(f"{number} is a prime number: {is_prime(number)}")
```

 Check whether given number is palindrome or not. 

```python
number = int(input("Enter a number: "))
def is_palindrome(n):
    original_number = str(n)
    reversed_number = original_number[::-1]
    return original_number == reversed_number
print(f"{number} is a palindrome: {is_palindrome(number)}")
```

Write a program to print sum of digits. 

```python
n = int(input("Enter a number: "))
def sum_of_digits(n):
    total = 0
    while n > 0:
        digit = n % 10
        total += digit
        n //= 10
    return total
print(f"The sum of the digits of {n} is: {sum_of_digits(n)}")
```

 Count and print all numbers divisible by 5 or 7 between 1 to 100.  

```python
for i in range(0, 101):
    if i % 5 == 0:
        print(i, "is divisible by 5")
    elif i % 7 == 0:
        print(i, "is divisible by 7")
```
