 Write a program to count and display the number of capital letters in a given 
string. 

```python
#wap to count and display the number of capital letters in a string
input_string = input("Enter a string: ")
def count_capital_letters(s):
    count = 0
    for char in s:
        if char.isupper():
            count += 1
    return count
print(f"Number of capital letters: {count_capital_letters(input_string)}",)
```

2. Count total number of vowels in a given string. 


```python
#count total number of vowels in a string
input_string = input("Enter a string: ")
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count
print(f"Number of vowels: {count_vowels(input_string)}")
```

WAP to enter a string and a substring. You have to print the number of times 
that the substring occurs in the given string. String traversal will take place from 
left to right, not from right to left.  

```python
#wap to enter a string and a substring and print the number of times the substring appears in the string, followed from left to right
input_string = input("Enter a string: ")
substring = input("Enter a substring: ")
def count_substring(s, sub):
    count = 0
    for i in range(len(s) - len(sub) + 1):
        if s[i:i+len(sub)] == sub:
            count += 1
    return count
print(f"Number of times the substring appears: {count_substring(input_string, substring)}")
```

5. Given a string containing both upper and lower-case alphabets. Write a Python 
program to count the number of occurrences of each alphabet (case 
insensitive) and display the same.  

```python
#count the number of occurrences of a specific character in a string, case sensitive<
input_string = input("Enter a string: ")
character = input("Enter a character: ")
def count_character(s, char):
    count = 0
    for c in s:
        if c == char:
            count += 1
    return count
print(f"Number of occurrences of the character: {count_character(input_string, character)}")
```

Program to count the number of unique words in a given sentence using sets. 

```python
#program to count thr number of unique words in a given sentence, using sets btw?
input_string = input("Enter a sentence: ")
def count_unique_words(s):
    words = s.split()
    unique_words = set(words)
    return len(unique_words)
print(f"Number of unique words: {count_unique_words(input_string)}")
```
