Q!:  Check whether a given number is divisible by 3 and 5 both. 

```python
#divisiblity checker
num = int(input("Enter a number: "))
if num%3 == 0:
    print(num, "is divisible by 3")
else:    print(num, "is not divisible by 3")
if num%5 == 0:
    print(num, "is divisible by 5")
else:    print(num, "is not divisible by 5")
if num%3 ==0 and num%5 == 0:
    print(num, "is divisible by both 3 and 5")
```

Find the greatest among two numbers. If numbers are equal than print 
“numbers are equal”.  

```python
a = int(input("Enter a number: "))
b = int(input("Enter another number: "))
if a > b:
    print(a, "is greater than", b)
elif a < b:
    print(a, "is less than", b)
elif a == b:
    print(a, "is equal to", b)
```

Check whether the quadratic equation has real roots or imaginary roots. 
Display the roots. 

```python
#real roots or imaginary roots
a = int(input("Enter a number: "))
b = int(input("Enter another number: "))
c = int(input("Enter a third number: "))
d = b**2 - 4*a*c
if d > 0:
    print("The equation has two real roots")
elif d == 0:
    print("The equation has one real root")
elif d < 0:
    print("The equation has imaginary roots")
```

Find whether a given year is a leap year or not. 

```python
#leap year checker
year = int(input("Enter a year: "))
if year%4 == 0 and year%100 != 0 or year%400 == 0:
    print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```
