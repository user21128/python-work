Write a program to compute the length of the hypotenuse (c) of a right triangle 
using Pythagoras theorem.  

```python
## h*h=b*b+p*p
base=float(input("Enter the base of the triangle: "))
perpendicular=float(input("Enter the perpendicular of the triangle: "))
hypotenuse=(base**2 + perpendicular**2)**0.5
print("The hypotenuse of the triangle is:", hypotenuse)
```

 Write a program to find sum of first n natural numbers.

 Write a program to print truth table for bitwise operators (&, |, and ^ operators) 

```python
#truth table of AND operator
print("A\tB\tA AND B")
print("0\t0\t", 0 and 0)
print("0\t1\t", 0 and 1)
print("1\t0\t", 1 and 0)
print("1\t1\t", 1 and 1)
#truth table of OR operator
print("A\tB\tA OR B")
print("0\t0\t", 0 or 0)
print("0\t1\t", 0 or 1)
print("1\t0\t", 1 or 0)
print("1\t1\t", 1 or 1)
#truth table of xor operator
print("A\tB\tA XOR B")
print("0\t0\t", 0 ^ 0)
print("0\t1\t", 0 ^ 1)
print("1\t0\t", 1 ^ 0)
print("1\t1\t", 1 ^ 1)

```

Write a program to find left shift and right shift values of a given number.


```python
#left shift operator
#right shift operator
a=int(input("Enter a number: "))
n=int(input("Enter the number of positions to shift: "))
left_shift=a<<n
right_shift=a>>n
print("Left shift of", a, "by", n, "positions is:", left_shift)
print("Right shift of", a, "by", n, "positions is:", right_shift)
```
