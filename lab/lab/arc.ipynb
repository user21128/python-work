```python
nums = [3, 0, 1]
def missing_number(nums):
    n = len(nums)
    total = n * (n + 1) // 2
    return int(total - sum(nums))
print(missing_number(nums))
```

```python
items = [("A", 90), ("B", 95), ("C", 95), ("D", 85)]
k = 2
def top_k(items, k):
    items.sort(key=lambda x: x[1], reverse=True)
    return items[:k]
print(top_k(items, k))
```

```python

def add_unique(x, seen=set()):
    if x not in seen:
        seen.add(x)
    return list(seen)
print(add_unique(1))
print(add_unique(2))
```

```python
pairs = [("sports","win"), ("sports","win"), ("tech","ai")]
def count_by_category(pairs):
    d = {}
    for cat, word in pairs:
        d.setdefault(cat, {})
        d[cat].setdefault(word, 0)
        d[cat][word] += 1
    return d
print(count_by_category(pairs))
```

```python
stream= [1, 2, 3, 4, 5]
w = 3

def moving_average(stream, w):
    window = []
    for x in stream:
        window.append(x)
        if len(window) > w:
            window.pop(0)
        yield sum(window) / len(window)

print(list(moving_average(stream, w)))
```

```python

def rle_decode(s):
    decoded = []
    for i in range(0, len(s), 2):
        count = int(s[i])
        char = s[i + 1]
        decoded.append(char * count)
    return ''.join(decoded)
```

```python
def is_balanced_quoted(s):
    in_quote = False
    for char in s:
        if char == '"':
            in_quote = not in_quote
    return not in_quote
print(is_balanced_quoted('She said, "Hello!"'))
```
