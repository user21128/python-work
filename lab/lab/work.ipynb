```python
class Counter:
    def __init__(self):
        self.value = 0

    def inc(self):
        self.value += 1

    def dec(self):
        self.value -= 1

    def show(self):
        return self.value
c = Counter()
c.inc()
```

```python
class Student:
    def set_details(self, name, roll):
        self.name = name
        self.roll = roll

    def display(self):
        print(self.name, self.roll)

s1 = Student()
s1.set_details("Aman", 101)
s2 = Student()
s2.set_details("Rohit", 102)
s3 = Student()
s3.set_details("Sneha", 103)
s3.display()
s2.display()
s1.display()
```
